# HuggingFace Sentiment Analysis (Colab Setup)
This notebook loads your CSV with customer feedback and applies a state-of-the-art transformer model to generate sentiment scores using HuggingFace.

```python
# Install dependencies
!pip install transformers torch pandas --quiet
```

```python
from transformers import pipeline
import pandas as pd
from google.colab import files
```

```python
# Upload CSV file
uploaded = files.upload()
```

```python
# Read uploaded file
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
df = df.dropna()
df['feedback'] = df['feedback'].str.strip()
df.head()
```

```python
# Load sentiment model
sentiment_pipeline = pipeline('sentiment-analysis')
```

```python
# Apply model to each feedback
def get_sentiment(text):
    result = sentiment_pipeline(text)[0]
    return pd.Series([result['label'], result['score']])

df[['hf_sentiment', 'hf_confidence']] = df['feedback'].apply(get_sentiment)
df.head()
```

```python
# Save to new CSV and download
df.to_csv('feedback_with_2_class_sentiment.csv', index=False)
files.download('feedback_with_2_class_sentiment.csv')
```
